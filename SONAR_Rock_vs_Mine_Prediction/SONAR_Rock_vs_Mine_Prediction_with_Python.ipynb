importing the dependencies 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score 
```

Data Collection and Data Procissing

```python
# loadig dataset 
sonar_data = pd.read_csv('/content/Copy of sonar data - Copy of sonar data.csv' , header = None)
```

```python
sonar_data.groupby(60).mean()
```

```python
# separating data and labels 
X = sonar_data.drop(columns= 60 , axis = 1 ) 
Y = sonar_data[60]
```

Training and Testing Data

```python
X_train , X_test , Y_train , Y_test = train_test_split(X ,Y , test_size = 0.2 , shuffle = True , random_state = 1 , stratify= Y)
```

Model_training -----> LogisticRegression

```python
model = LogisticRegression()
```

```python
model.fit(X_train , Y_train )
```

Model Evaluation

```python
# Accuracy of training data 
train_data_prediction = model.predict(X_train)
train_data_accuracy = accuracy_score(train_data_prediction , Y_train)
```

```python
print('accuracy of training data ' + str(train_data_accuracy * 100) + ' %')
```

```python
#Accuracy of Testing data 
test_data_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(test_data_prediction , Y_test)
```

```python
print('accuracy of testing data ' + str(test_data_accuracy * 100) + ' %')
```
